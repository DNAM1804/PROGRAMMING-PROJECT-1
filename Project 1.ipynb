## DATA CORRELATION

```python
fa_values = data.groupby('fixed acidity').count().sort_values(by ='quality',ascending= False)["quality"][:7.3]
```

```python
fa_values
```

```python
axes = fa_values.plot.bar(rot=0,stacked=True,color="green")
plt.title('HIGHEST ELEVEN FIXED ACIDITY VALUES IN TERMS OF QUALITY')
```

```python
fa_values.count()
```

**This bar graph shows the relationship between some fixed acidity values and the quality in descending order.We see that, as the quality of the red decreases , there is a rise and fall of the value of fixed acidity in which the difference between values of the fixed acid column are almost insignificant.**

```python
sns.relplot(data, x= 'density',y='residual sugar',color= 'brown', s=40)
plt.savefig('image2.png')
plt.title('RELATIONSHIP BETWEEN ALCOHOL AND RESIDUAL SUGAR')
plt.show()
```

**From this graph,we see that there is not much relationship between alcohol and residual sugar because ponts are scattered and are not positioned in any proper form.**

```python
data.corr()
```

**From the dataset given here, we can see the connection between the various factors that affect the quality of wine with other factors.The table shows a sort of interconnection between the factors.Correlation between attributes and themselves gives 1.At the poinnts where there are negatives, it shows that there is a negative correlation (connection) between the factors.On the last row, we see that the quality of the red wine is greatly affected by the alcohol and not really affected by the pH, residual sugar, density,chlorides,free and total sulfur dioxide .**

```python
sns.relplot(data, x= 'free sulfur dioxide',y='total sulfur dioxide',color= 'red', s=30)
plt.title('CORRELATION BETWEEN FREE SULFUR DIOXIDE AND TOTAL SULFUR DIOXIDE')
plt.show()
```

**This shows a low positive corerlation between free and total sulfur dioxide since the points seem to be concentrated at a point but seems to be following a straight line.** 

```python
sns.relplot(data, x= 'fixed acidity',y='volatile acidity',color= 'grey', s=40)
plt.savefig('image2.png')
plt.title('RELATIONSHIP BETWEEN FIXED ACIDITY AND VOLATILE ACIDITY')
plt.show()
```

**From the scatter plot,we can deduce that there is no correlation as the points plotted are scattered and follow no definite pattern.**

```python
sns.relplot(data, x= 'quality',y='alcohol',color= 'black', s=30)
plt.title('RELATIONSHIP BETWEEN QUALITY AND ALCOHOL')
plt.show()
```

**Here, both variables alcohol and quality are continuous,discrete and dense.**

```python
sns.heatmap(data.corr(),annot =True) 
plt.figure(figsize=(20,10))
plt.savefig('image3.png')
plt.show()
```

**A heat map shows the relationship between two variables, one plotted on the x and the other on the y-axis. From our heat map, we can tell how one factor is changing with respect to the other factors.It is almost similar to the data correlation function.The only difference is that we use the warmness or the coolness of color to ahiw how it is changing with respect to the other factors. Here the factors against themselves is shown by theplain white boxes.The lower the value, the darker the colour becomes. From the heatmap, we can tell that alcohol has a lot of influence on the quality of the red wine since it has a light color and which means its value in relation to the quality is high.Other hidden inferences can be shown on the heat map.**
